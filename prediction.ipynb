```python
import pandas as pd
```

```python
model=load_model('model.h5')
```

```python
with open('onehotencoder.pkl','rb')as file:
    geo=pickle.load(file) 
```

```python
with open('scaler.pkl','rb') as f:
    sc=pickle.load(f)
```

```python
with open('label_encoder_gender.pkl','rb') as f:
    l=pickle.load(f)
    
```

```python
input_data={
  'CreditScore': 600,
  'Geography': 'France',
  'Gender': 'Female',
  'Age': 40,
  'Tenure': 3,
  'Balance': 60000,
  'NumOfProducts': 2,
  'HasCrCard': 1, 
  'IsActiveMember': 1,
  'EstimatedSalary': 50000
}
```

```python
da=pd.DataFrame{(input_data),index=[0]}
```

```python
geo_encoded=geo.transform([[input_data['Geography']]]).toarray()
```

```python
geo_df=pd.DataFrame(geo_encoded,columns=geo.get_feature_names_out(
['Geography']
))
```

```python
input_df=pd.DataFrame([input_data])
```

```python
input_df
```

```python
input_data=pd.concat([input_df.drop('Geography',axis=1),geo_df],axis=1)
```

```python
input_data
```

```python
input_data['Gender']=l.transform(input_data['Gender'])
```

```python
input_data
```

```python
input_scaled=sc.transform(input_data)
```

```python
input_scaled 
```

```python
pred=model.predict(input_scaled)
```
